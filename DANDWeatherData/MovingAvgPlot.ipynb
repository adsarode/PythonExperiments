```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

```python
global_data = pd.read_csv("Global.csv")
global_data.head()
```

```python
pune_data = pd.read_csv("Pune.csv")
pune_data.head()
```

```python
pune_data['MAPune'] = pune_data['avg_temp'].rolling(window=10).mean()
pune_data.head()
```

```python
plt.style.use('ggplot')
plt.plot(pune_data['MAPune'])
plt.plot(pune_data['avg_temp'])
```

```python
global_data['MAGlobal'] = global_data['avg_temp'].rolling(window=10).mean()
global_data.head()
```

```python
plt.style.use('ggplot')
plt.plot(pune_data['year'], pune_data['MAPune'], color = 'r', label = 'Pune',linewidth=2.0)
plt.plot(global_data['year'], global_data['MAGlobal'], color = 'b', label = 'Global',linewidth=2.0)

plt.xlabel('Year', fontsize = 30)
plt.ylabel('Temperature', fontsize = 30)

plt.legend(loc=2, prop={'size': 25})

plt.title('Compare Global vs Pune Templerature Changes', fontsize = 40)
matplotlib.rc('figure', figsize=(20, 10))

plt.show()
```
