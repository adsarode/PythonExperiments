# matplotlib

* Library used under the hood by pandas
* Low level

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
xs = np.linspace(-6,6,100)
ys = np.sinc(xs)
```

```python
df = pd.DataFrame({'x':xs, 'sinc':ys})
```

```python
df.plot.line(x='x', y='sinc')
```

# Style

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
df.plot.line(x='x',y='sinc')
```

```python
plt.style.use('seaborn-paper')
df.plot.line(x='x',y='sinc')
```

```python
plt.style.use('seaborn-white')
df.plot.line(x='x',y='sinc')
```
